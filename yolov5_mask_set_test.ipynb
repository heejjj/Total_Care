```python
# %cat -> 데이터를 출력해라 
%cat /content/dataset/data.yaml

#여기 Train과 val 경로를 수정해야함 
# Traine데이터 셋 안의 모든 이미지 데이터를 취합해서 하나의 파일로 만들기
# 이미지 경로를 따내는 방법은 -> from glob import glob 경로들을 텍스트로 변환해 줌 
```

```python
from glob import glob


img_list = glob('/content/dataset/train/images/*.jpg')
val_list = glob('/content/dataset/valid/images/*.jpg')
print(img_list[:5])
#이미지의 경로가 들어 있음
print(len(img_list))
print(len(val_list))
```

```python
#경로가 들어간 img_list, val_list를 txt 파일로 저장하기 

with open('/content/dataset/train.txt', 'w') as f:
  f.write('\n'.join(img_list)+'\n')

with open('/content/dataset/val.txt', 'w') as f:
  f.write('\n'.join(img_list)+'\n')
```

```python
%cat /content/dataset/data.yaml


#yaml데이터에 txt파일의 경로를 알려줘야 한다. 
#yaml은 일반 텍스트파일이 아니라서 import yaml가져와야함
```

```python
import yaml
with  open('/content/dataset/data.yaml', 'r',) as f:
 data = yaml.safe_load(f)

 print(data)

 #data에 yaml이 딕셔너리 형태로 들어가짐 
```

```python
data.keys()

#train과 val의 경로를 각각 train.txt 위치로 수정해야함
```

```python
data['train'] = '/content/dataset/train.txt'
data['val'] = '/content/dataset/val.txt'

data
```

```python
#위의 data변수를 yaml.dump()를 써서 저장
#data.yaml에 'w'(write)함 

with open ('/content/dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)

%cat /content/dataset/data.yaml

#경로가 잘 바뀜
```
